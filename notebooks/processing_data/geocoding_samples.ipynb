# Geocoding samples to prepare for large scale geocoding

Likely, there will be many repeat locations in our geocoding task and geocoding API calls are expensive. We will cache results in a local DB to re-use geocoding results when possible. We will host cached geocoded results in a PostGIS DB locally (in a `geocoding_cache` table).

At this stage of processing, we will assume names are already normalized, and that if a name is provided, it is expected to be in our look-up or new.
